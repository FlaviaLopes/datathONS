```python
import os
import unidecode
import re
from pdfminer.high_level import extract_text
from glob import glob
from datetime import datetime
import pandas as pd
import numpy as np
from difflib import SequenceMatcher
from fuzzywuzzy import fuzz
```

```python
class Extrator:
  def __init__(self, file):
    self.file = file
    self.text = self.extract_from_pdf()

  def extract_from_pdf(self) -> str:
    # TODO: tem parte não sendo extraídas
    # exemplo pdf 9 (com problemas) n extra se ribeiro gonçalves, legenda de tabela.
    try:
      return extract_text(self.file)
    except ValueError:
      return ''
    
  def to_ascii(self):
    '''
    padroniza o texto [...]
    '''
    outstring = ''
    for i in range(len(self.text)):
      outstring = outstring + unidecode.unidecode(self.text[i])
    self.text = outstring

  def clean_text(self):
    '''
    transforma em lowercase e remove quebras de linha e outros caracteres
    '''
    self.text = str.lower(self.text)
    self.text = re.sub('_', '', self.text)
    self.text = re.sub('\n', '', self.text)
    self.text = re.sub(' +', ' ', self.text)

class Validador:
  def __init__(self, text, name_file, base, base_agentes):
    self.text = text
    self.name_file = name_file
    self.base = pd.read_excel(base)
    self.base_agentes = pd.read_csv(base_agentes, sep=';', dtype=str)
    self.base_agentes.columns = self.base_agentes.columns.str.replace(' ', '')


  def __extract_parecer(self) -> str:
      '''
      extrai do documento o código do Parecer de Acesso de acordo com um padrão regex
      '''
      pa_pattern = '[0-9]{4}[-]pa[-][0-9]{4}'
      pa = np.unique(re.findall(pa_pattern, self.text))
      return pa[0] if pa.size else ''    
      
  def __has_parecer(self, parecer: str) -> bool:
    '''
    verifica se o parecer extraído do documento está cadastrado na base
    '''
    return self.base.documento_emitido.str.contains(parecer).any()

  def __valida_cnpjs(self):
      CNPJ_ONS = '02.831.210/0001-57'
      CNPJ_PATTERN = '[0-9]{2}.[0-9]{3}.[0-9]{3}\/[0-9]{4}\-[0-9]{2}'

      def loc_cnpjs(text):
        cnpjs = re.findall(pattern=CNPJ_PATTERN, string=text.replace(' ', ''))
        return cnpjs
      
      def check_cnpj(cnpj):
        cnpj = ''.join(re.findall(pattern='\d+', string=cnpj))
        
        if len(self.base_agentes.query(f"DscCnpj == '{cnpj}'")):
            return True
        else:
            return False

      cnpjs = loc_cnpjs(self.text)
      
      val_cnpjs = {'transmissora': {'cnpj': [], 'validado':False},
                  'usuario': {'cnpj': [], 'validado':False},
                  'ons': {'cnpj': [], 'validado':False},
                  'anuentes': {'cnpj': [], 'validado':[]}}

      cnpjs_checked = []
      
      for i, cnpj in enumerate(cnpjs):
          if cnpj not in cnpjs_checked:
              if i == 0:
                  val_cnpjs['transmissora'] = {'cnpj': cnpj, 'validado': check_cnpj(cnpj)}
              elif i == 1:
                  val_cnpjs['usuario'] = {'cnpj': cnpj, 'validado': check_cnpj(cnpj)}
              elif i == 2 or cnpj == CNPJ_ONS:
                  if cnpjs[2] == CNPJ_ONS:
                      val_cnpjs['ons'] = {'cnpj': cnpj, 'validado': True}
                  else:
                      val_cnpjs[cnpj] = {'cnpj': cnpj, 'validado': False}
              else:
                  val_cnpjs['anuentes']['cnpj'].append(cnpj)
                  val_cnpjs['anuentes']['validado'].append(check_cnpj(cnpj))
              cnpjs_checked.append(cnpj)
      return val_cnpjs
  
  def __to_ascii(self, ls):
    outstring = ''
    for i in range(len(ls)):
      outstring = outstring + unidecode.unidecode(ls[i])
    return outstring

  def __clean_text(self, text):
    text = str.lower(self.__to_ascii(text))
    pattern=r'[^a-zA-z0-9\s]'
    text=re.sub(pattern,'',text)
    text=re.sub('_', '', text)
    text=re.sub('\n', '', text)
    text=re.sub(' +', ' ', text)
    return text
  
  def __find_pattern(self, text, pattern):
    i = 0
    n = 1
    interesse = []
    while n > 0:
      n = text.find(pattern,i)
      if n > 0:
        interesse.append(n)
      i = n + 1
    return interesse
  
  def __validate_string_cnpj(self, string2check, clean_text, num_cnpj): #num_cnpj é uma string somente de números, sem marcas
    ra_back = 500           # configuração
    ra_foward = 200  
    valido = False          # inicialização
    ratio = 100 
    if self.__find_pattern(clean_text, num_cnpj):
      pos_cnpj = self.__find_pattern(clean_text, num_cnpj)[0]  # posição que aparece o CNPJ
      texto_amostra = clean_text[max(pos_cnpj-ra_back,0):pos_cnpj+ra_foward] # amostra ao redor do CNPJ
      
      if string2check in texto_amostra:
        valido = True
      
      else:        
        alternativa = string2check

        #%%
        # Tentar localizar parte intermediárias da string
        if alternativa.count(" ") > 4: # remove pedaços da string
          alternativa = alternativa.split(' ', 1)[1]

        if alternativa.count(" ") > 3:
          alternativa = alternativa.rsplit(' ', 1)[0] # reduz a string padrão
        
        #%%

        extra = len(string2check) - len(alternativa)
        texto_amostra_ampliado = clean_text[max(pos_cnpj-2*ra_back,0):pos_cnpj+2*ra_foward] # aumenta o tamanho da amostra
        
        # nova busca do padrão em espaço maior
        mark = self.__find_pattern(texto_amostra_ampliado, alternativa) # amplia o espaço de busca / posição do texto no texto ampliado

        if mark != []:
          # candidato = clean_text[max(pos_cnpj-2*ra_back,0):pos_cnpj+2*ra_foward] # amostra com o cnpj no meio \ mesmo que texto_amostra_ampliado

          candidato = texto_amostra_ampliado[max(mark[0]-extra,0):mark[0]+extra+len(alternativa)] # ajusta para minimizar os caracteres adicionais
          print('<------------')
          print('Texto Buscado ('+ num_cnpj +'): ' + string2check)
          print('Candidato encontrado no documento: ' + candidato) 

          # Comparação das strings
          try: 
            ratio = fuzz.ratio(candidato, string2check) #* ((extra + len(razao[i-k]))/len(razao[i-k])) ajuste não se mostrou necessário      
          except:
            ratio = -1
          
          if ratio > 70:
            valido = True
            print('Ratio ok: ' + str(ratio))

          else:
            valido = False
            print('Ratio baixo: ' + str(ratio))
          
          print('------------>')  

        else:
          valido = False
          ratio = -1
    else:
      ratio = -1

    return valido, ratio
  
  def __check_name_agente(self, cnpj):     #num_cnpj é uma string somente de números, sem marcas
    text = self.text
    
    if text != '':
      texto_limpo = self.__clean_text(text)
      num_cnpj = ''.join(re.findall(pattern='\d+', string=cnpj))
      nome = self.__clean_text(self.base_agentes[self.base_agentes['DscCnpj'] == num_cnpj]['DscRazaoSocial'].iloc[0])
      valido, ratio = self.__validate_string_cnpj(nome, texto_limpo, num_cnpj)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]

  def __check_street_agente_address(self, cnpj):     #num_cnpj é uma string somente de números, sem marcas
    aux = self.base_agentes
    text = self.text
    
    if text != '':
      texto_limpo = self.__clean_text(text)
      num_cnpj = ''.join(re.findall(pattern='\d+', string=cnpj))
      nome = self.__clean_text(aux[aux['DscCnpj'] == num_cnpj]['DscLogradouro'].iloc[0])
      valido, ratio = self.__validate_string_cnpj(nome, texto_limpo, num_cnpj)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]
  
  def __check_city_agente_address(self, cnpj):     #num_cnpj é uma string somente de números, sem marcas
    aux = self.base_agentes
    text = self.text
    
    if text != '':
      texto_limpo = self.__clean_text(text)
      num_cnpj = ''.join(re.findall(pattern='\d+', string=cnpj))
      nome = self.__clean_text(aux[aux['DscCnpj'] == num_cnpj]['Nome'].iloc[0])
      valido, ratio = self.__validate_string_cnpj(nome, texto_limpo, num_cnpj)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]
  
  def __check_name_ons(self):
    cnpj_ons = '02831210000157'
    razao_ons = 'operador nacional do sistema eletrico ons'
    if self.text != '':
      texto_limpo = self.__clean_text(self.text)
      valido, ratio = self.__validate_string_cnpj(razao_ons, texto_limpo, cnpj_ons)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]
  
  def __check_street_ons_brasilia(self):   
    cnpj_ons = '02831210000157'  
    street_ons = 'asa sul area de servicos publicos'
    if self.text != '':
      texto_limpo = self.__clean_text(self.text)
      valido, ratio = self.__validate_string_cnpj(street_ons, texto_limpo, cnpj_ons)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]
  
  def __check_street_ons_rio(self):   
    cnpj_ons = '02831210000157'  
    street_ons = 'rua julio do carmo'
    if self.text != '':
      texto_limpo = self.__clean_text(self.text)
      valido, ratio = self.__validate_string_cnpj(street_ons, texto_limpo, cnpj_ons)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]
  
  def __check_cidade_ons_brasilia(self):   
    cnpj_ons = '02831210000157'  
    street_ons = 'brasilia'
    if self.text != '':
      texto_limpo = self.__clean_text(self.text)
      valido, ratio = self.__validate_string_cnpj(street_ons, texto_limpo, cnpj_ons)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]
  
  def __check_cidade_ons_rio(self):   
    cnpj_ons = '02831210000157'  
    street_ons = 'rio de janeiro'
    if self.text != '':
      texto_limpo = self.__clean_text(self.text)
      valido, ratio = self.__validate_string_cnpj(street_ons, texto_limpo, cnpj_ons)
    else:
      valido = False
      ratio = -1
    return [valido, ratio]
  
  

  def item1(self):
    try:
      c = self.__valida_cnpjs()
      status = 'OK' if c['ons']['validado'] else 'FALHA'
      return (status, '', '')   
    except:
      return ('NAO_VALIDADO', '', '') 
  
  def item2(self):
    try:
      c = self.__valida_cnpjs()
      status = 'OK' if c['transmissora']['validado'] else 'FALHA'
      return (status, '', '') 
    except:
      return ('NAO_VALIDADO', '', '') 

  def item3(self):
    try:
      c = self.__valida_cnpjs()
      status = 'OK' if c['usuario']['validado'] else 'FALHA'
      return (status, '', '') 
    except:
      return ('NAO_VALIDADO', '', '') 

  def item4(self):
    try:
      c = self.__valida_cnpjs()
      if (c['anuentes']['validado'] == []) | np.array(c['anuentes']['validado']).all():
        status = 'OK'
      else:
        status = 'FALHA'
      return (status, '', '') 
    except:
      return ('NAO_VALIDADO', '', '') 

  def item5(self):
    try:
      name_ons = self.__check_name_ons()
      if name_ons[0]:
        cidade_brasilia = self.__check_cidade_ons_brasilia()
        cidade_rio = self.__check_cidade_ons_rio()
        if cidade_rio[0] or cidade_brasilia[0]:
          street_rio = self.__check_street_ons_rio()
          street_brasilia = self.__check_street_ons_brasilia()
          if street_rio[0] or cidade_brasilia[0]:
            status = 'OK'
          else:
            status = 'FALHA'
        else:
          status = 'FALHA'
      else:
        status = 'FALHA'
        
      return (status, '', '')
    except:
      return ('NAO_VALIDADO', '', '') 

  def item6(self):
    try:
      agente = 'transmissora'
      cnpjs = self.__valida_cnpjs()[agente]
      
      if cnpjs['validado']:
        name_agente = self.__check_name_agente(cnpjs['cnpj'])
        city_agente = self.__check_city_agente_address(cnpjs['cnpj'])
        street_agente = self.__check_street_agente_address( cnpjs['cnpj'])
        ratio = (name_agente[1] + city_agente[1] + street_agente[1])/3
        
        if name_agente[0] + city_agente[0] + street_agente[0] == 3:
          
          status = 'OK'
        else:
          status = 'FALHA'
        
      else:
        status = 'FALHA'
        ratio = -1
        
      return (status, ratio, '')
    except:
      return ('NAO_VALIDADO', '', '') 

  def item7(self):
    try:
      agente = 'usuario'
      cnpjs = self.__valida_cnpjs()[agente]
      
      if cnpjs['validado']:
        name_agente = self.__check_name_agente(cnpjs['cnpj'])
        city_agente = self.__check_city_agente_address(cnpjs['cnpj'])
        street_agente = self.__check_street_agente_address( cnpjs['cnpj'])
        ratio = (name_agente[1] + city_agente[1] + street_agente[1])/3
        
        if name_agente[0] + city_agente[0] + street_agente[0] == 3:
          status = 'OK'
        else:
          status = 'FALHA'
        
      else:
        status = 'FALHA'
        ratio = -1
        
      return (status, ratio, '')
    except:
      return ('NAO_VALIDADO', '', '') 

  def item8(self):
    try:
      agente = 'anuentes'
      cnpjs = self.__valida_cnpjs()[agente]
      
      if len(cnpjs['cnpj']):
        for cnpj in cnpjs['cnpj']:
          name_agente = self.__check_name_agente(cnpj)
          city_agente = self.__check_city_agente_address(cnpj)
          street_agente = self.__check_street_agente_address(cnpj)
          ratio = (name_agente[1] + city_agente[1] + street_agente[1])/3
          
          if name_agente[0] + city_agente[0] + street_agente[0] == 3:
            status = 'OK'
          else:
            status = 'FALHA'
            break
      
      else:
        status = 'OK'
        ratio = ''
        
      return (status, ratio, '')
    except:
      return ('NAO_VALIDADO', '', '') 

  def item9(self):
    def similar(a, b):
        return SequenceMatcher(None, a, b).ratio()

    def check_contrato(contratos, registros):        
        max_similaridades = {}        
        for j, contrato in enumerate(contratos):
            contrato = contrato.split('/')
            contrato = f'{contrato[1]}-{contrato[0]}'
            
            for i, registro in enumerate(registros):
                registro = registro[4:]
                if i == 0 and j == 0:
                    max_similaridades = {'contrato': contrato, 'registro': registro, 'razao': similar(contrato, registro)}
                else:
                    if max_similaridades['razao'] < similar(contrato, registro):
                        max_similaridades = {'contrato': contrato, 'registro': registro, 'razao': similar(contrato, registro)}
                
        return max_similaridades

    def valida_contrato_concessao(texto):
        cnpjs = self.__valida_cnpjs()
        
        if cnpjs['transmissora']['validado']:
            cnpj_transmissora = cnpjs['transmissora']['cnpj']
            cnpj_transmissora = ''.join(re.findall(pattern='\d+', string=cnpj_transmissora))
            registros = list(set(self.base_agentes.query(f'DscCnpj == "{cnpj_transmissora}"')['CodRegulamentacao'].to_list()))
            #reg = re.findall(pattern='[concessão|portaria].*nº[0-9]{3}/[0-9]{4}', string=texto.replace(' ', ''))[0]
            reg = re.findall(pattern='n[o|º][0-9]{3}/[0-9]{4}', string=e.text.replace(' ', ''))[0]
            contratos_concessao_possiveis = re.findall(pattern='[0-9]{3}/[0-9]{4}', string=reg)
            
            similaridades = check_contrato(contratos_concessao_possiveis, registros)
        
        else:
            similaridades = {'contrato': False, 'registro': False, 'razao': 0}
            
        status = 'OK' if similaridades['razao'] > 0.7 else 'FALHA'
        return (status, similaridades['razao'], '')
    try: 
      return valida_contrato_concessao(self.text)
    except:
      return ('NAO_VALIDADO', '', '') 

  def item10(self):
    return ('NAO_VALIDADO', '', '') 

  def item11(self):
    try:
      pa = re.findall('parecer de acesso', self.text)
      status = 'OK' if pa else 'FALHA'
      return (status, '', '') 
    except:
      return ('NAO_VALIDADO', '', '') 

  def item12(self):    
    def get_parecer_date(parecer: str):
      '''
      recupera da base a data do parecer que está no documento
      '''
      return self.base.loc[self.base.documento_emitido.str.contains(parecer), 'data_emissao'].values[0]

    try:
      aviso = ''
      pa = self.__extract_parecer()
      if self.__has_parecer(pa):
        date = get_parecer_date(pa)
        days = (datetime.now().date() - pd.to_datetime(date).date()).days
        if days >= 90:
          aviso = f'parecer de acesso vencido: {days} dias.'
        elif days >= 60 & days < 90:
          aviso = f'parecer de acesso perto de vencer: faltam {90 - item12} dias.'
      else:
        aviso = f'parecer de acesso não encontrado na base: {pa}.'    
      return ('AVISO', '', aviso)
    except:
      return ('NAO_VALIDADO', '', '') 

  def item13(self):
    def get_ponto_conexao(parecer: str) -> str:
      ponto = self.base.loc[self.base.documento_emitido.str.contains(parecer), 'pontoconexao'].values
      if ponto.size:
        return ponto[0]
      else:
        return ''
    try:
      aviso = ''
      pa = self.__extract_parecer()
      if self.__has_parecer(pa):
        pc = get_ponto_conexao(pa)
        if pc in self.text:
          aviso = f'PC {pc} do PA está presente no CCT.'
        else:
          aviso = 'PC do PA ausente no CCT.'
        if pc == '':
          aviso = 'PC não encontrado para o PA (ausente na base).'
      else:
        aviso = f'parecer de acesso não encontrado na base: {pa}.'
      return ('AVISO', '', aviso)
    except:
      return ('NAO_VALIDADO', '', '') 

  def item14(self):
    try:
      forma_invalida = 'diretor geral e diretor de administração de servicos de transmissão'
      qualificacao_diretores = self.text.replace(',', '')\
                                    .replace('  ', ' ')\
                                    .split("02.831.210/0001-57")[1]\
                                    .split('representado')[1]\
                                    .split(';')[0]
      razao = fuzz.ratio(forma_invalida, qualificacao_diretores)
      
      if razao >= 0.7:
          return ('FALHA', razao, '')
      else:
          return ('OK', razao, '')
    except:
      return ('NAO_VALIDADO', '', '') 

  def item15(self):
    DEFINICOES_VALIDAS = {
    "procedimentos de rede": "documento elaborado pelo ons com a participação dos agentes que, aprovado pela aneel, estabelece os procedimento e requisitos técnicos necessários para o planejamento, para a implantação, para o uso e para a operação do sin, bem como as responsabilidades do ons e dos agentes.",
    "regras de transmissão": "documentos de cárater normativo que estabelecem os requisitos técnicos necessários para a operação, das instalações de trasmissão, bem como as atividades de supervisão, coordenação e controle do sin."
    }

    try:
        procedimento_doc = self.text.replace(' ', '').split('"procedimentosderede":')[1].split('.')[0].split(';')[0]
        razao_procedimento_rede = fuzz.ratio(DEFINICOES_VALIDAS['procedimentos de rede'].replace(' ', ''), procedimento_doc)
        razao_procedimento_transmissao = fuzz.ratio(DEFINICOES_VALIDAS["regras de transmissão"].replace(' ', ''), procedimento_doc)
        
        razao = max([razao_procedimento_rede, razao_procedimento_transmissao])

        if razao >= 70:
            resposta = ('OK', razao, '')
        elif 50 < razao < 70:
            resposta = ('AVISO', razao, 'Descrição não satisfatória')
        else:
            resposta = ('FALHA', razao, '')
    except:
        resposta = ('FALHA', 0, 'Procedimentos de rede não encontrados')
    
    return resposta

  def item16(self):
    try:
      # "procedimentos de rede":
        procedimento_doc = self.text.replace(' ', '').split('"procedimentosderede":')[1].split('.')[0].split(';')[0]
        
        if len(procedimento_doc):
          resposta = ('OK', '', '')
        
        else:
          resposta = ('AVISO', '', 'Procedimentos de rede não encontrado')
    except:
        resposta = ('AVISO', 0, 'Procedimentos de rede não encontrados')
    
    return resposta

  def item17(self):
    return ('NAO_VALIDADO', '', '')

  def item18(self):
    return ('NAO_VALIDADO', '', '')

  def item19(self):
    return ('NAO_VALIDADO', '', '')

  def item20(self):
    return ('NAO_VALIDADO', '', '')

  def item21(self):
    return ('NAO_VALIDADO', '', '')

  def item22(self):
    return ('NAO_VALIDADO', '', '')

  def item23(self):
    return ('NAO_VALIDADO', '', '')

  def item24(self):
    '''
    Esta função retorna True caso seja encontrado o seguinte erro:
         Erro! Fonte de referência não encontrada
    Caso ela retorne True o usuário deve receber um alerta informando que este erro foi encontrado
    '''
    erros_ref = re.findall(pattern='erro!fontedereferencianaoencontrada', string=self.text.replace(' ', ''))
    
    if len(erros_ref):
        return ('AVISO', '', 'Verificar Fonte de Referência')
    else:
        return('OK', '', '')
  
  def item25(self):
    '''
    Esta função alerta se há menção a data de entrade em operação|antecipação no documento.
    Caso seja encontrada a menção ela retorna True e caso contrário retorna False.
    Ainda será necessário verificar o contrato a função retorne Falso.
    '''
    ocorrencias = [m.start() for m in re.finditer('data de entrada em [operacao|antecipacao]', self.text)]
    
    if len(ocorrencias):
        return ('AVISO', '', 'Menção a data de operação encontrada')
    else:
        return ('OK', '', 'Não foi encontrada mensão a data de operação')
    
  def validacao(self):
    output = {self.name_file: {}}    
    metodos = [self.item1, self.item2, self.item3, self.item4, self.item5, 
               self.item6, self.item7, self.item8, self.item9, self.item10,
               self.item11, self.item12, self.item13, self.item14, self.item15,
               self.item16, self.item17, self.item18, self.item19, self.item20,
               self.item21, self.item22, self.item23, self.item24, self.item25]

    for i in range(25):
      status, grau, trecho = metodos[i]()
      output[self.name_file].update({
          i + 1: {'STATUS': status, 'GRAU_CONFIANCA': grau, 'TRECHO': trecho}
      })
    return output
```
